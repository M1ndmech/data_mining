```python
fill_median = ["LotFrontage", "MasVnrArea"]
fill_mode = ["GarageYrBlt"]
fill_none = ["BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "Electrical", "FireplaceQu", "GarageType", "GarageFinish", "GarageQual", "GarageCond"]
```

```python
train_data[fill_median] = train_data.groupby("Neighborhood")[fill_median].transform(lambda x: x.fillna(x.median()))
train_data[fill_mode] = train_data.groupby("Neighborhood")[fill_mode].transform(lambda x: x.fillna(x.mode()[0]))
```

```python
train_data[fill_none] = train_data[fill_none].fillna("None")
```

### Проведите разведочный анализ данных (EDA), ответив на следующие вопросы: 
a. Каково распределение целевой переменной 'SalePrice'? Есть ли какие-либо выбросы? 

```python
def normal(mean, std, histmax=False, color="black"):
    x = np.linspace(mean-4*std, mean+4*std, 200)
    p = stats.norm.pdf(x, mean, std)
    if histmax:
        p = p*histmax/max(p)
    z = plt.plot(x, p, color, linewidth=2)
```

```python
sales_plot_data = train_data['SalePrice']

data = np.random.normal(size=500) * 0.1    
ax = sns.histplot(x=sales_plot_data, stat="probability")
normal(sales_plot_data.mean(), sales_plot_data.std(), histmax=ax.get_ylim()[1])
```

Распределение не нормальное, есть выбросы за зоной статистической значимости (цена выше ~700000)
